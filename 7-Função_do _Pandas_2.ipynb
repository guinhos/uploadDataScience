## Agregações conplexas em Pandas (GroupBy: Split, Apply, Combine)

#### Muitas vezes, é importante poder fazer operações de [agregação](https://towardsdatascience.com/pandas-groupby-aggregate-transform-filter-c95ba3444bbb) de forma condicional a algum subconjunto de dados (por exemplo, para casos que atendam a alguma condição). Isso é implementado pelo operador [`groupby`](https://towardsdatascience.com/how-to-use-groupby-and-aggregate-functions-in-pandas-for-quick-data-analysis-c19e7ea76367)

####  O nome [`groupby`](https://pandas.pydata.org/docs/getting_started/comparison/comparison_with_sql.html#group-by) é proveniente da linguagem SQL.

#### É útil pensar sobre os termos de [Hadley Wickham (SAC): split, apply, combine](https://medium.com/analytics-vidhya/split-apply-combine-strategy-for-data-mining-4fd6e2a0cc99).

### Split, apply, combine:

- A etapa [`split`](https://medium.com/@sureshssarda/pandas-splitting-exploding-a-column-into-multiple-rows-b1b1d59ea12e) envolve dividir e reagrupar um ``DataFrame`` baseado em uma determinada chave.
- A etapa [`apply`](https://towardsdatascience.com/introduction-to-pandas-apply-applymap-and-map-5d3e044e93ff) pressupõe calcular alguma função (geralmente, alguma agregação, transformação ou filtro) sobre os grupos constituídos na etapa anterior.
- A etapa [`combine`](https://medium.com/dev-genius/combining-data-in-pandas-31c984afceb7) faz uma "mesclagem" dos resultados dessas operações em um novo arranjo.

#### Embora cada uma das etapas possa ser feita "manualmente", o [`GroupBy`](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html) geralmente pode fazê-las em uma única etapa:

####  A vantagem do `GroupBy` é que ele permite abstrair as três etapas anteriores: o usuário não precisa pensar em "como" fazer o cálculo, mas na operação como um todo.

#### A operação [`split`-`apply`-`combine`](https://towardsdatascience.com/how-to-use-the-split-apply-combine-strategy-in-pandas-groupby-29e0eb44b62e) mais básica com o método `groupby()` é passar o nome de uma chave de coluna específica.

### O objeto `GroupBy`.

#### O objeto [`.groupBy()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html) é uma [abstração](https://medium.com/@lucasoliveiras/descomplicando-o-groupby-f57ff2dcd6db) muito flexível. Pode ser tratado como uma coleção de `DataFrames` e talvez as operações mais importantes de um objeto [`GroupBy`](https://naomi-fridman.medium.com/pandas-groupby-explained-with-titanic-6a1c47eb8182) sejam:

- aggregate; 
- filter;
- transform; 
- apply.

```python
cwur.groupby('country')
```

#### Observe que o método [.groupby()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html?highlight=groupby#pandas-dataframe-groupby) não retorna um `DataFrame`, mas um objeto [`DataFrameGroupBy`](https://pandas.pydata.org/pandas-docs/stable/reference/groupby.html). Podemos pensar no `DataFrameGroupBy` como uma visualização especial de um `DataFrame` que constrói os grupos, mas não executa nenhum cálculo até que o estágio de agregação seja aplicado.

#### Esta avaliação faz com que as operações de agregação possam ser calculadas de forma muito eficiente. Para produzir um resultado, podemos aplicar uma [função de agregação](https://realpython.com/pandas-groupby/) a este `DataFrameGroupBy` para executar a operação (aplicar) e prosseguir na etapa de combinação.

```python
passo_1 = cwur.groupby('country', as_index = False).sum()

passo_2 = passo_1.sort_values("citations", ascending = False)

passo_3 = passo_2.head()

passo_4 = passo_3.loc[:, ['country', 'world_rank']]

passo_4
```

#### Da para  agrupar  o `DataFrame` `cwur` pela coluna `country`, somando o número de universidades por país, ordenar pelo número de citações de forma descendente e mostrar apenas as primeiras linhas desse sub `DataFrame`.

```python
#cwur
cwur.groupby('country', as_index = False).sum().sort_values("citations", ascending = False).head()
```

```python
cwur.groupby('country', as_index=False).sum().sort_values("citations", ascending = False).head().loc[:,['country','world_rank']]
```

Como antes, nenhum cálculo é feito até que alguma [função de agregação](https://cmdlinetips.com/2019/10/pandas-groupby-13-functions-to-aggregate/#:~:text=Pandas%20groupby%20function%20enables%20us,dimension%20of%20the%20grouped%20object.) sobre o objeto seja invocada.

### Iteração sobre grupos:

#### O objeto `GroupBy` suporta a iteração direta sobre grupos, retornando cada grupo como uma `Series` ou um `DataFrame`.

```python
lista = [1, 2, 3, 4]

for i in lista:
    print(i)
```

```python
for (pais, group) in cwur.groupby('country'):
    if group.citations.sum() < 5000:
        print("{0:20s} citacoes = {1:20f} ".format(pais, group.citations.sum()))
```

#### A seguir vamos agregar os dados de mínimo, máximo, média e soma as medida de `"score"` por instituição.

```python
cwur.groupby(['country', 'year'])[['score', 'citations']].aggregate([min, np.median, max, sum]).head()
```

#### Outra maneira útil é usar um dicionário que mapeia nomes de coluna com operações. Desta forma, a operação é aplicada a cada coluna.

```python
cwur.groupby(['country', 'year'])[['score', 'citations']].aggregate({'score':np.mean, 'citations':sum}).head()
```

### Filtering

#### Uma operação de [filtragem](https://cmdlinetips.com/2018/02/how-to-subset-pandas-dataframe-based-on-values-of-a-column/) permite "descartar" dados com base nas propriedades do grupo. Veja a seguir a quantidade de citações que cada país possui, assim como a quantidade de instituições.

```python
above_50k = cwur.groupby('country').aggregate(['sum', 'count'])['citations'].sort_values('sum', ascending = False)
above_50k
```

#### Filtrar as instituições de países que possuem um número de citações maior ou igual a um determinado valor de corte. Considerando apenas países com mais de 50.000 citações, teríamos EUA, China, Japão e França.

```python
above_50k_passo_1 = cwur.groupby('country').aggregate(['sum', 
                                                       'count'])['citations'].sort_values('sum', 
                                                                                          ascending = False).head(4)
print(above_50k_passo_1)
print('\nTotal de universidades em países com mais de 50k citações:', above_50k_passo_1['count'].sum())
```

#### Assim denifini uma função de filtragem para o número de citações.

```python
def filter_func(x):
    return x['citations'].sum() > 50000

cwur.groupby('country').filter(filter_func)
```

#### A função filtro retorna um booleano especificando se o grupo passa pelo filtro ou não.

### Transformation:

#### Enquanto a agregação retorna uma versão reduzida dos dados, a [transformação](https://www.analyticsvidhya.com/blog/2020/03/understanding-transform-function-python/) retorna alguma versão transformada dos dados para, então, fazer a combinação. A saída de uma transformação tem o mesmo `shape` que a entrada.

#### O método [`.transform()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.transform.html) chama uma função "on self", produzindo um Data Frame com valores [transformados](https://morioh.com/p/bfd87b86e9fe) pela função chamada.

```python
brazil = cwur[cwur.country == "Brazil"]
brazil
```

```python
#brazil.publications - brazil.publications.mean()
brazil['publications'] - brazil['publications'].mean()
```

#### Função para calcular o desvio da posição central da distribuição.

```python
def center_mean(x):
    return(x - x.mean())

center_mean_output = cwur[["country", "publications"]].groupby('country').transform(center_mean)
print(center_mean_output.loc[330,:])
```

#### Aqui é aplicado o método [`.apply()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.groupby.GroupBy.apply.html) juntamente com o método [`.groupby()`](https://towardsdatascience.com/learn-how-to-master-groupby-function-in-python-now-4620dd463224) e o resultado é uma coluna de citações normalizada pela número total de publicações da instituição.

```python
def norm_by_data2(x):
    '''
     x is a DataFrame of group values
    '''
#    x['publications'] /= x['citations'].sum()
    x['citations'] /= x['publications'].sum()
    return x

cwur[['country', 
      'publications', 
      'citations', 
      'institution']].groupby('country').apply(norm_by_data2)
```

####  |Definir uma das colunas como índice de um novo dataframe e criar um dicionário das cinco primeiras universidades rankeadas, como a soma de diversas quantidades.

```python
cwur2 = cwur.drop(['world_rank', 
                   'national_rank', 
                   'year', 
                   'quality_of_education', 
                   'quality_of_faculty', 
                   'influence'
                  ], 
                  axis = 1
                 )
cwur2 = cwur2.set_index('institution')
cwur2
mapping = {'Harvard University': 'HU',
           'Massachusetts Institute of Technology': 'MIT',
           'Stanford University': 'SU',
           'University of Cambridge': 'UoC',
           'California Institute of Technology': 'Calteh'
          }
cwur2.groupby(mapping).sum()
```

#### Uma lista de chaves válidas pode ser agrupada com base em [multi-indexadores](https://towardsdatascience.com/handling-pandas-groupby-and-its-multi-indexes-efae3e6b788c).
